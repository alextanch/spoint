```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
%load_ext autoreload
%autoreload 2

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

from spoint.utils.primitives import Primitives
from spoint.utils.augmentation import Augmentation, Resize
from spoint.utils.config import Config

def draw_image(image):
    plt.imshow(image, cmap='gray')
    plt.show()
```

## Render primitive

```python
image, points = Primitives(['lines', 'polygon', 'ellipses', 'star', 'stripes', 'cube', 'checkerboard'])()
image, points = Resize(config.size)(image, points)

draw_image(image)
```

## Test gaussian noise

```python
config = Config({
    'size': (120, 160),
    'gaussian_noise': {'prob': 1, 'stddev': [10, 20]}
})

aug_image = Augmentation(config)(image)
draw_image(aug_image)
```

## Test speckle noise

```python
config = Config({
    'size': (120, 160),
    'speckle_noise': {'prob': 1, 'range': [.0, .005]}
})

aug_image = Augmentation(config)(image)
draw_image(aug_image)
```

## Test random brightness

```python
config = Config({
    'size': (120, 160),
    'random_brightness': {'prob': 1, 'max_change': 100}
})

aug_image = Augmentation(config)(image)
draw_image(aug_image)
```

## Test random contrast

```python
config = Config({
    'size': (120, 160),
    'random_contrast': {'prob': 1, 'range': (0.5, 1.5)}
})

aug_image = Augmentation(config)(image)
draw_image(image)
draw_image(aug_image)
```

## Test additive shade

```python
config = Config({
    'size': (120, 160),
    'additive_shade': {'prob': 1, 'ellipses': 20, 'transparency': [-0.5, 0.8], 'ksize': [250, 350]}
})

aug_image = Augmentation(config)(image)
draw_image(image)
draw_image(aug_image)
```

## Test motion blur

```python
config = Config({
    'size': (120, 160),
    'motion_blur': {'prob': 1, 'ksize': 20}
})

aug_image = Augmentation(config)(image)
draw_image(image)
draw_image(aug_image)
```

## Test random augmentations

```python
config = Config({
    'size': (120, 160),
    'gaussian_noise': {'prob': 0.5, 'stddev': [1, 10]},
    'speckle_noise': {'prob': 0.5, 'range': [.0, .005]},
    'random_brightness': {'prob': 0.5, 'max_change': 100},
    'random_contrast': {'prob': 0.5, 'range': (0.5, 1.5)},
    'additive_shade': {'prob': 0.5, 'ellipses': 20, 'transparency': [-0.5, 0.8], 'ksize': [250, 350]},
    'motion_blur': {'prob': 0.5, 'ksize': 20}
})

aug_image = Augmentation(config)(image)
draw_image(image)
draw_image(aug_image)
```
