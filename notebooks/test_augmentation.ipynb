```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
%load_ext autoreload
%autoreload 2

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

from spoint.utils.primitives import Primitives
from spoint.utils.augmentation import Augmentation
from spoint.utils.config import Config

def draw_image(image):
    plt.imshow(image, cmap='gray')
    plt.show()
```

## Render primitive

```python
image, points = Primitives(['lines', 'polygon', 'ellipses', 'star', 'stripes', 'cube', 'checkerboard'])()
draw_image(image)
```

## Test gaussian noise

```python
config = Config({
    'size': (120, 160),
    'gaussian_noise': {'prob': 1, 'stddev': [10, 20]}
})

aug_image = Augmentation(config, size=config.size)(image)
draw_image(aug_image)
```

## Test speckle niose

```python
config = Config({
    'size': (120, 160),
    'speckle_noise': {'prob': 1, 'stddev': [.0, .005]}
})

aug_image = Augmentation(config, size=config.size)(image)
draw_image(aug_image)
```

## Test random brightness

```python
config = Config({
    'size': (120, 160),
    'random_brightness': {'prob': 1, 'max_change': 100}
})

aug_image = Augmentation(config, size=config.size)(image)
draw_image(aug_image)
```
