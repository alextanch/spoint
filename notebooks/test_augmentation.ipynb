```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
%load_ext autoreload
%autoreload 2

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

from spoint.utils.primitives import Primitives
from spoint.utils.augmentation import Augmentation, Resize
from spoint.utils.config import Config

def draw_image(image):
    plt.imshow(image, cmap='gray')
    plt.show()
```

## Render primitive

```python
config = Config({
    'background': {'size': [600, 800]},
    'primitives': ['lines', 'polygon', 'ellipses', 'star', 'stripes', 'cube', 'checkerboard']
})

image, points = Primitives(config)()

draw_image(image)
```

## Test gaussian noise

```python
config = Config({
    'gaussian_noise': {'prob': 1, 'stddev': [10, 20]}
})

aug_image = Augmentation(config)(image)
draw_image(aug_image)
```

## Test speckle noise

```python
config = Config({
    'speckle_noise': {'prob': 1, 'range': [.0, .005]}
})

aug_image = Augmentation(config)(image)
draw_image(aug_image)
```

## Test random brightness

```python
config = Config({
    'random_brightness': {'prob': 1, 'max_change': 100}
})

aug_image = Augmentation(config)(image)
draw_image(aug_image)
```

## Test random contrast

```python
config = Config({
    'random_contrast': {'prob': 1, 'range': (0.5, 1.5)}
})

aug_image = Augmentation(config)(image)
draw_image(image)
draw_image(aug_image)
```

## Test additive shade

```python
config = Config({
    'additive_shade': {'prob': 1, 'ellipses': 10, 'transparency': [0.8, 1], 'ksize': [50, 100]}
})

aug_image = Augmentation(config)(image)
draw_image(image)
draw_image(aug_image)
```

## Test motion blur

```python
config = Config({
    'size': (120, 160),
    'motion_blur': {'prob': 1, 'ksize': 40}
})

aug_image = Augmentation(config)(image)
draw_image(image)
draw_image(aug_image)
```

## Test random augmentations

```python
config = Config({
    'random_brightness': {'prob': 1, 'max_change': 100},
    'random_contrast': {'prob': 1, 'range': (0.5, 1.5)},
    'additive_shade': {'prob': 1, 'ellipses': 20, 'transparency': [0.9, 1], 'ksize': [50, 100]},
    'motion_blur': {'prob': 1, 'ksize': 20},
    'gaussian_noise': {'prob': 1, 'stddev': [1, 10]},
    'speckle_noise': {'prob': 1, 'range': [.0, .005]}
})

aug_image = Augmentation(config)(image)
draw_image(image)
draw_image(aug_image)
```
