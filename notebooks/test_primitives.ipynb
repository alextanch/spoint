```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
%load_ext autoreload
%autoreload 2

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

from spoint.utils.primitives import Primitives

def draw_points(image, points):
    image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    for p in points:
        image = cv2.circle(image, (p[1], p[0]), 16, (255,0,0), -1) 
    
    plt.imshow(image)
    plt.show()
```

## Test draw background

```python
from spoint.utils.primitives import Background

image = Background()()

plt.imshow(image, cmap='gray')
plt.show()
```

## Test draw lines

```python
image, points = Primitives(['lines'])()
draw_points(image, points)
```

## Test polygon

```python
image, points = Primitives(['polygon'])()
draw_points(image, points)
```

## Test ellipses

```python
image, points = Primitives(['ellipses'])()
draw_points(image, points)
```

## Test star

```python
image, points = Primitives(['star'])()
draw_points(image, points)
```

## Test stripes

```python
image, points = Primitives(['stripes'])()
draw_points(image, points)
```

## Test cube

```python
image, points = Primitives(['cube'])()
draw_points(image, points)
```

## Test checkerboard

```python
image, points = Primitives(['checkerboard'])()
draw_points(image, points)
```

## Test random primitive

```python
image, points = Primitives(['lines', 'polygon', 'ellipses', 'star', 'stripes', 'cube', 'checkerboard'])()
draw_points(image, points)
```
