```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
%load_ext autoreload
%autoreload 2

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

from spoint.utils.primitives import Primitives
from spoint.utils.config import Config

def draw_points(image, points):
    image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    for p in points:
        image = cv2.circle(image, (p[0], p[1]), 6, (255,0,0), -1) 
    
    plt.imshow(image)
    plt.show()
```

## Test draw background

```python
from spoint.utils.primitives import Background

config = Config({
    'background': {'size': [600, 800]},
})

image = Background()(**config.background)

plt.imshow(image, cmap='gray')
plt.show()
```

## Test draw lines

```python
config = Config({
    'background': {'size': [600, 800]},
    'primitives': ['lines']
})

image, points = Primitives(config)()
draw_points(image, points)
```

## Test polygon

```python
config = Config({
    'background': {'size': [600, 800]},
    'primitives': ['polygon']
})

image, points = Primitives(config)()
draw_points(image, points)
```
